**Esercizio 4:**

provare ad implementare un esperimento toy per un'algoritmo basilare di regressione lineare

1.   viene dato il seguente dataset di coppie di misure $T\{x_i,y_i\}, i=1,\cdots N$:



```
x = [0., 2.22, 4.44, 6.67, 8.89, 11.11, 13.33, 15.56, 17.788, 20.0]
```

2.  si ipotizza che il campione sia stato prodotto in accordo ad una funzione lineare del tipo: $y = \theta x + \epsilon\,,\quad \epsilon\sim\mathcal N(0,\sigma)$, in cui $x\in\mathbb{R}$ rappresenta l'input, $\theta=0.6$, e $y\in\mathbb{R}$ le osservazioni "rumorose" ($\epsilon$ è un numero random gaussiano estratto evento per evento)

3. si chiede di scrivere ed allenare un modello lineare di ML che apprenda la funzione di mapping $y=f_\theta(x)$, addestrando l'algoritmo in modo empirico in base a $k$ step di addestramento, in cui, partendo da un valore random del parametro $\theta$ nell'intervallo $(0,1)$, si aggiorni il valore di $\theta$ ad ogni step muovendosi in un intervallo piccolo intorno al valore corrente del paramento, e accettando la nuova soluzione in accordo alla funzione di costo $MSE$ definita come: $MSE(\theta) = \frac{1}{N}\sum_{i=1}^N (y_i - f_\theta(x_i))^2$, in cui la somma è fatta su gli elementi del campione.

3.   testare le predizioni del modello con input $x \in [0.,20.]$ e confrontare con l'andamento teorico ipotizzato ($\theta = 0.6$), calcolando l'MSE tra risultati predetti e quelli veri attesi assumendo $\theta=0.6$.

4.   guardare alla soluzione proposta per vedere come è possibile graficare in un plot $y$ VS $x$, i risultati ottenuti, sfruttando la libreria python **matplotlib** (che vedremo più in dettaglio nella prossima lezione).

````
# Esempi di addestramento
x = np.array([0., 2.22, 4.44, 6.67, 8.89, 11.11, 13.33, 15.56, 17.788, 20.0])
y = np.array([-2.1489976, 1.26684956, 5.00441023, 3.84220594, 5.95983884, 6.44189515, 6.88238407, 12.04053287, 10.73255652, 10.29414663])
theta_true = 0.6

import random
theta = random.random() # inizializzo in modo random in (0,1) il parametro da addestrare
k = 15 # step di addestramento
lr = 0.001 # learning rate

def function(x, theta):  # funzione con la quale genero le y partendo da theta e x
  return theta*x

def mse_loss(y_pred, y_true):  # funzione che calcola la loss MSE
  return (1/ len(y_true)) * np.sum((y_true - y_pred) ** 2)

def theta_update(x, y, y_pred, lr, theta):  # aggiornamento del parametro theta tramite gradient descent
  gradient = (2/len(x)) * np.sum(x * (y_pred - y))  # calcola il gradiente
  return theta - lr*gradient  # aggiornamento classico della discesa del gradiente

loss_hist = []
theta_hist = []

# training loop
for epoch in range(k):
    y_pred = function(x, theta)  # predizione
    loss = mse_loss(y_pred, y)  # calcolo della loss
    theta = theta_update(x, y, y_pred, lr, theta)  # aggiornamento di theta

    loss_hist.append(loss)  # salvataggio
    theta_hist.append(theta)

    # salvataggio dei parametri migliori
    if epoch == 0:
      best_loss = loss
      best_theta = theta
    else:
      if loss < best_loss:
        best_loss = loss
        best_theta = theta

    loss_prev = loss  # aggiornamento della loss precedente
    print(f'epoch {epoch} \t loss: {loss:.3f} \t theta: {theta:.3f}')

print(f'\nbest loss: {best_loss:.3f} for best theta: {best_theta:.3f}')

import matplotlib.pyplot as plt
epochs = np.arange(1, k+1, 1)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # crea un'unica figura con 3 sottografi affiancati

# Primo grafico
axes[0].plot(theta_hist, loss_hist)
axes[0].set_title('mse vs theta')
axes[0].set_xlabel('theta')
axes[0].set_ylabel('mse loss')
axes[0].grid()

# Secondo grafico
axes[1].plot(epochs, loss_hist)
axes[1].set_xticks(epochs)
axes[1].set_title('mse vs epochs')
axes[1].set_xlabel('epochs')
axes[1].set_ylabel('mse loss')
axes[1].grid()

# Terzo grafico
axes[2].plot(x, y, linestyle='', marker='x', label='data')
axes[2].plot(x, function(x, best_theta), label='pred')
axes[2].plot(x, function(x, theta_true), label='true')
axes[2].set_title('ground truth & prediction')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
axes[2].legend()
axes[2].grid()

plt.tight_layout()  # Per evitare sovrapposizioni
plt.show()
````

**Esercizio suggerito:**

1. estendere il modello al caso lineare a due parametri: intercetta e slope: $w + \theta*x$

````
# Esempi di addestramento
x = np.array([0., 2.22, 4.44, 6.67, 8.89, 11.11, 13.33, 15.56, 17.788, 20.0])
y = np.array([-2.1489976, 1.26684956, 5.00441023, 3.84220594, 5.95983884, 6.44189515, 6.88238407, 12.04053287, 10.73255652, 10.29414663])
theta_true = 0.6

import random
theta = random.random() # inizializzo in modo random in (0,1) il parametro da addestrare
w = np.random.uniform(-1, 1)  # inizializzo l'intercetta nell'intervallo [-1,1]
k = 30 # step di addestramento
lr = 0.001 # learning rate

def function2(x, theta, w):  # funzione con la quale genero le y partendo da theta e x
  return theta*x + w

def mse_loss(y_pred, y_true):  # funzione che calcola la loss MSE
  return (1/ len(y_true)) * np.sum((y_true - y_pred) ** 2)

def param_update(x, y, y_pred, lr, theta, w):  # aggiornamento del parametro theta tramite gradient descent
  gradient = (2/len(x)) * np.sum(x * (y_pred - y))  # calcola il gradiente
  theta -= lr*gradient  # aggiornamento classico della discesa del gradiente
  w -= lr*gradient
  return theta, w

loss_hist = []
theta_hist = []

# training loop
for epoch in range(k):
    y_pred = function2(x, theta, w)  # predizione
    loss = mse_loss(y_pred, y)  # calcolo della loss
    theta, w = param_update(x, y, y_pred, lr, theta, w)  # aggiornamento di theta

    loss_hist.append(loss)  # salvataggio
    theta_hist.append(theta)

    # salvataggio dei parametri migliori
    if epoch == 0:
      best_loss = loss
      best_theta = theta
      best_w = w
    else:
      if loss < best_loss:
        best_loss = loss
        best_theta = theta
        best_w = w

    loss_prev = loss  # aggiornamento della loss precedente
    print(f'epoch {epoch} \t loss: {loss:.3f} \t theta: {theta:.3f} \t w: {w:.3f}')

print(f'\nbest loss: {best_loss:.3f} for best theta: {best_theta:.3f} & best w: {best_w:.3f}')

import matplotlib.pyplot as plt
epochs = np.arange(1, k+1, 1)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # crea un'unica figura con 3 sottografi affiancati

# Primo grafico
axes[0].plot(theta_hist, loss_hist)
axes[0].set_title('mse vs theta')
axes[0].set_xlabel('theta')
axes[0].set_ylabel('mse loss')
axes[0].grid()

# Secondo grafico
axes[1].plot(epochs, loss_hist)
axes[1].set_xticks(np.arange(epochs.min(), epochs.max()+1, step=2))
axes[1].set_title('mse vs epochs')
axes[1].set_xlabel('epochs')
axes[1].set_ylabel('mse loss')
axes[1].grid()

# Terzo grafico
axes[2].plot(x, y, linestyle='', marker='x', label='data')
axes[2].plot(x, function2(x, best_theta, best_w), label='pred')
axes[2].plot(x, function(x, theta_true), label='true')
axes[2].set_title('ground truth & prediction')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
axes[2].legend()
axes[2].grid()

plt.tight_layout()  # Per evitare sovrapposizioni
plt.show()
````
